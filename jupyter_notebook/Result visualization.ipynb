```python
import pandas as pd
import os
import json
```

```python
file_list = ['abalone', 'heart', 'tic-tac-toe']
missingness_flag = [0, 10, 20, 30, 40, 50]  
clf_list = ['DecisionTreeClassifier', 'KNeighborsClassifier', 'RandomForestClassifier', 'XGBClassifier']
imputation_list = ['GAIN', 'KNNImputer', 'MICE', 'SimpleImputer_constant', 'SimpleImputer_mean'
                  , 'SimpleImputer_median', 'SimpleImputer_most_frequent']
```

```python
result_path = "C:\\Users\\Administrator\\Desktop\\Code_Test\\Research\\Data-Imputation-master\\result"
```

```python
def txt_extraction(path):
    txt_file = open(path, "r")
    txt = txt_file.read()
    #Covert string to dictionary
    result_dict = json.loads(txt)
    return result_dict['mean']
```

```python
def result_extraction(file_name_path, imputation_list, missingness, clf, method):
#     data_missing = {'DecisionTreeClassifier':[],
#                     'KNeighborsClassifier':[],
#                     'RandomForestClassifier':[],
#                     'XGBClassifier':[]
#                    }
    results = []
    for imp_algo in imputation_list:
        clf_folder_path = os.path.join(file_name_path, '{}/data_missing_{}'.format(imp_algo, missingness))
        result_path = os.path.join(clf_folder_path, clf)
        rs_path = os.path.join(result_path, '{}.txt'.format(method))
        result = txt_extraction(rs_path)
        results.append(result)
    return results
```

```python
from IPython.core import display as ICD
```

```python
def visualization_data(file_name, missingness):
    print('File name: {} - missingness: {}'.format(file_name, missingness))
    file_name_path = os.path.join(result_path, file_name)
    data_missing = {'DecisionTreeClassifier':None,
                    'KNeighborsClassifier':None,
                    'RandomForestClassifier':None,
                    'XGBClassifier':None
                   }
    for clf in clf_list:
        results = result_extraction(file_name_path, imputation_list, missingness, clf, 'accuracy')
        data_missing['{}'.format(clf)] = pd.Series(results, index = imputation_list)
    return data_missing
```
